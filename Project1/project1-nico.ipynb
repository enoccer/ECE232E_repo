We observe a Gaussian distribution that starts centered at low degree values near the origin (when the probability p is low) and gradually shifts to the right (when p grows).

Since the degree distribution follows a binomial distribution, the expected value (peak) has the relationship n * p, where n is the number of nodes. Since n is fixed, an increased p value would shift the center of the distribution progressively to the right, which is observed in our plots.

A binomial distribution increasingly resembles a Gaussian distribution if n * p is sufficiently large. The rule of thumb is often n * p >= 5 and n * (1-p) >= 5. We observe our distributions resemble a Gaussian one as p increases.
