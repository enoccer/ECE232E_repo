A binomial distribution increasingly resembles a Gaussian distribution if n * p is sufficiently large. The rule of thumb is often n * p >= 5 and n * (1-p) >= 5. We observe our distributions resemble a Gaussian one as p increases.

#### (c)

The general PageRank equation 
$$\pi(i) = (1 - \alpha) \sum_{j=1}^{n} \left( \frac{1}{k_{out}(j)} A_{ji} \pi(j) \right) + \frac{\alpha}{n}$$
transforms to
$$\pi(i) = (1 - \alpha) \sum_{j=1}^{n} \left( \frac{1}{k_{out}(j)} A_{ji} \pi(j) \right) + \frac{\alpha}{|T|}$$,
where $T$ is a set of trusted nodes. If the node $i$ is not a trusted node, the personalized PageRank equation reduces to 
$$\pi(i) = (1 - \alpha) \sum_{j=1}^{n} \left( \frac{1}{k_{out}(j)} A_{ji} \pi(j) \right)$$
